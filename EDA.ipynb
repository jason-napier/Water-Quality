```python
#Functions

#Function using us to convert State Postal Codes into State Names.
def get_state_name(state_code):
    state = us.states.lookup(state_code)
    if state is not None:
        return state.name
    else:
        return None
```

```python
# Find the columns for 2020
education_2020_df = combined_df.loc[:,["State","Area name", "Less than a high school diploma, 2017-21", "High school diploma only, 2017-21", "Some college or associate's degree, 2017-21", "Bachelor's degree or higher, 2017-21", "Percent of adults with less than a high school diploma, 2017-21", "Percent of adults with a high school diploma only, 2017-21", "Percent of adults completing some college or associate's degree, 2017-21", "Percent of adults with a bachelor's degree or higher, 2017-21"]]

#Convert State code to State Name
education_2020_df['State'] = education_2020_df['State'].astype(str).apply(get_state_name)

#Groupby state
education_2020_df = education_2020_df.groupby("State").mean()

education_2020_df.head()
```

```python
# Group by State
# Load your dataframe into a variable called `df`
df = pd.read_csv("my_data.csv")

# Group the data by "State" column and compute the mean for each column
means_by_state = df.groupby("State").mean()

# Loop through each column and print the name of the column and its mean value by state
for col in means_by_state.columns:
    print("Column name:", col)
    print(means_by_state[col])
```

```python
#Create new dataframe with means by state
# Create an empty DataFrame to store the results
state_mean_df = pd.DataFrame(columns=['State', 'MeanValue'])

# Iterate over the unique states in the original DataFrame
for state in df['State'].unique():
    
    # Calculate the mean value for the current state
    mean_value = df.loc[df['State'] == state, 'Value'].mean()
    
    # Add the state and mean value to the results DataFrame
    state_mean_df = state_mean_df.append({'State': state, 'MeanValue': mean_value}, ignore_index=True)
    
# Save the results to a CSV file
state_mean_df.to_csv('state_means.csv', index=False)
```
