```python
#Functions

#Function using us to convert State Postal Codes into State Names.
def get_state_name(state_code):
    state = us.states.lookup(state_code)
    if state is not None:
        return state.name
    else:
        return None
```

```python
# Find the columns for 2020
education_2020_df = combined_df.loc[:,["State","Area name", "Less than a high school diploma, 2017-21", "High school diploma only, 2017-21", "Some college or associate's degree, 2017-21", "Bachelor's degree or higher, 2017-21", "Percent of adults with less than a high school diploma, 2017-21", "Percent of adults with a high school diploma only, 2017-21", "Percent of adults completing some college or associate's degree, 2017-21", "Percent of adults with a bachelor's degree or higher, 2017-21"]]

#Convert State code to State Name
education_2020_df['State'] = education_2020_df['State'].astype(str).apply(get_state_name)

#Groupby state
education_2020_df = education_2020_df.groupby("State").mean()

education_2020_df.head()
```

```python
#Find Columns for 2020
population_2020_df = combined_df.loc[:,["State","Area name","Population 2020"]]

#Change state codes to Names
population_2020_df['State'] = population_2020_df['State'].astype(str).apply(get_state_name)

#Group by State
population_2020_df = population_2020_df.groupby("State").mean()

population_2020_df.head()
```

```python
#Create Dataframes for water quality

#Uranium
#import file
uranium_df = pd.read_csv("Resources\CDC\ANNUAL PERCENT OF CWS BY MEAN URANIUM CONCENTRAtion (MGL).csv")

#Rename
uranium_df = uranium_df.rename(columns= {"Value":"Percent by Mean Uranium Concentration >30 mg/L"})
uranium_df.head()
```

```python
#arsenic
#import file
arsenic_df = pd.read_csv("Resources\CDC\ARSENIC IN COMMUNITY WATER SYSTEMS (CWS)  ANNUAL PERCENT OF CWS BY MEAN ARSENIC CONCENTation (MG per L).csv")

#Rename
arsenic_df = arsenic_df.rename(columns= {"Value":"Percent by Mean Arsenic Concentration >10 mg/L"})
arsenic_df.head()
```

```python
#DEHP
#import file
dehp_df = pd.read_csv("Resources\CDC\DI(2-ETHYLHEXYL) PHTHALATE (DEHP).csv")

#Rename
dehp_df = dehp_df.rename(columns= {"Value":"Percent by Mean DI(2-ETHYLHEXYL) PHTHALATE (DEHP) Concentration > 6 mg/L"})
dehp_df.head()
```

```python
#Haloacetics Acids
#import file
hha_df = pd.read_csv("Resources\CDC\DISINFECTION BYPRODUCTS IN COMMUNITY WATER SYSTEMS (CWS)  ANNUAL PERCENT OF CWS BY MEAN Haloacetics Acids Concentration.csv")

#Rename
hha_df = hha_df.rename(columns= {"Value":"Percent by Mean Haloacetics Acids Concentration >60 mg/L"})
hha_df.head()
```

```python
#Nitrate
#import file
nitrate_df = pd.read_csv(r"Resources\CDC\NITRATES IN COMMUNITY WATER SYSTEMS (CWS) ANNUAL PERCENT.csv")

#Rename
nitrate_df = nitrate_df.rename(columns= {"Value":"Percent by Mean Nitrate Concentration >10 mg/L"})
nitrate_df.head()
```

```python
#Merge to create impurities_df
temp_df = pd.merge(uranium_df, arsenic_df, on="State", how="outer")
temp_df = pd.merge(temp_df, dehp_df, on="State", how="outer")
# temp_df = pd.merge(temp_df, hha_df, on="State", how="outer")
temp_df = pd.merge(temp_df, nitrate_df, on="State", how="outer")
temp_df.head()
# temp_df.columns
```

```python
impurities_df = temp_df.loc[:,["State","Year_x","Percent by Mean Uranium Concentration >30 mg/L","Percent by Mean Arsenic Concentration >10 mg/L","Percent by Mean DI(2-ETHYLHEXYL) PHTHALATE (DEHP) Concentration > 6 mg/L","Percent by Mean Nitrate Concentration >10 mg/L"]]

impurities_df = impurities_df.rename(columns={"Year_x":"Year"})
# impurities_df.drop(impurities_df[2])
impurities_df
```

```python
#List of States
# Define the mapping between states and regions
state_to_region = {
    'Arizona': 'Four Corners',
    'California': 'West Coast',
    'Colorado': 'Four Corners',
    'Connecticut': 'Northeast',
    'Maine': 'Northeast',
    # 'Maryland': 'South',
    'Massachusetts': 'Northeast',
    'New Hampshire': 'Northeast',
    'New Jersey': 'Northeast',
    'New Mexico': 'Four Corners',
    'New York': 'Northeast',
    'Oregon': 'West Coast',
    'Pennsylvania': 'Northeast',
    'Rhode Island': 'Northeast',
    'Utah': 'Four Corners',
    'Vermont': 'Northeast',
    'Washington': 'West Coast'
}

# Apply the mapping to the "State" column and create a new "Region" column
impurities_df['Region'] = impurities_df['State'].map(state_to_region)

impurities_region_df = impurities_df.groupby("Region").mean()
impurities_region_df
```

```python
#Combine Census data together
#Dataframes to combine for the year 2020
census_2020_df = pd.merge(population_2020_df, unemployment_2020_df, on="State", how="outer")
census_2020_df = pd.merge(census_2020_df, education_2020_df, on="State", how="outer")
census_2020_df = pd.merge(census_2020_df, poverty_2020_df, on="State", how="outer")

#Reset index to create "State" column
census_2020_df.reset_index()

census_2020_df.info()
```

```python
#Merge impurities_df and Census 2020 dataframe
complete_2020_df =pd.merge(census_2020_df, impurities_df, on="State",how="outer")
complete_2020_df.head()
```

```python
# Group by State
# Load your dataframe into a variable called `df`
df = pd.read_csv("my_data.csv")

# Group the data by "State" column and compute the mean for each column
means_by_state = df.groupby("State").mean()

# Loop through each column and print the name of the column and its mean value by state
for col in means_by_state.columns:
    print("Column name:", col)
    print(means_by_state[col])
```

```python
#Create new dataframe with means by state
# Create an empty DataFrame to store the results
state_mean_df = pd.DataFrame(columns=['State', 'MeanValue'])

# Iterate over the unique states in the original DataFrame
for state in df['State'].unique():
    
    # Calculate the mean value for the current state
    mean_value = df.loc[df['State'] == state, 'Value'].mean()
    
    # Add the state and mean value to the results DataFrame
    state_mean_df = state_mean_df.append({'State': state, 'MeanValue': mean_value}, ignore_index=True)
    
# Save the results to a CSV file
state_mean_df.to_csv('state_means.csv', index=False)
```
