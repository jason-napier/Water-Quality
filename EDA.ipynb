Program for loading csv files and performing exploratory data analysis

```python
#Imports
import pandas as pd
import us
```

```python
#Functions

#Function using us library to convert State Postal Codes into State Names.
def get_state_name(state_code):
    state = us.states.lookup(state_code)
    if state is not None:
        return state.name
    else:
        return None
```

```python
#Dataframes of Note:

# combined_df           #Dataframe combining all Census data
# education_2020_df     #Eduction statistic for 2020
# unemployment_2020_df  #Unemployment statstics for 2020
# population_2020_df    #Population statistics for 2020
# poverty_2020_df       #Poverty rates for 2020
# impurities_df         #Dataframe merging datframes of impurities
# uranium_df            #Uranium Water Quality for 2020
# arsenic_df            #Arsenic Water Quality for 2020
# dehp_df               #DI(2-ETHYLHEXYL) PHTHALATE (DEHP)
#hha_df                 #Haloacetics Acids
#nitrate_df             #Nitrate
```

```python
# Load data files

#Education
education_df = pd.read_excel("Resources\Education_modified.xlsx")
education_df.head()
```

```python
#Population
population_df = pd.read_excel("Resources\PopulationEstimates_modified.xlsx")
population_df.head()
```

```python
#Poverty Estimates
poverty_df = pd.read_excel("Resources\PovertyEstimates_modified.xlsx")
poverty_df.head()
```

```python
#Unemployment
unemployment_df = pd.read_excel("Resources/Unemployment_modified.xlsx")
unemployment_df.head()
```

```python
#Merge Education and Population dataframes
merge_1_df = pd.merge(education_df, population_df, on=["State", "Area name"], how="outer")
merge_1_df.info()
```

```python
#Merge merge_1_df dataframe and Unemployment dataframe
merge_2_df = pd.merge(merge_1_df,unemployment_df, left_on=["State","Area name"], right_on=["State","Area_name"], how="outer")
merge_2_df.info()
```

```python
#Merge merged_2_df with Poverty dataframe
merge_3_df = pd.merge(merge_2_df,poverty_df, left_on=["State","Area name"], right_on=["Stabr","Area_name"], how="outer")
merge_3_df.info()
```

```python
#If user wishes to add additional dataframes, please combine them here with the previous dataframe 
```

```python
#Rename dataframe after merges are complete
combined_df = merge_3_df.copy()
```

```python
# Find the columns for 2020 for Education
education_2020_df = combined_df.loc[:,["State","Area name", "Less than a high school diploma, 2017-21", "High school diploma only, 2017-21", "Some college or associate's degree, 2017-21", "Bachelor's degree or higher, 2017-21", "Percent of adults with less than a high school diploma, 2017-21", "Percent of adults with a high school diploma only, 2017-21", "Percent of adults completing some college or associate's degree, 2017-21", "Percent of adults with a bachelor's degree or higher, 2017-21"]]

#Convert State code to State Name
education_2020_df['State'] = education_2020_df['State'].astype(str).apply(get_state_name)

#Groupby state
education_2020_df = education_2020_df.groupby("State").mean()

education_2020_df.head()
```

```python
#Find Columns for 2020 for Unemployment
unemployment_2020_df = combined_df.loc[:,["State","Area name", "Civilian_labor_force_2020", "Employed_2020", "Unemployed_2020", "Unemployment_rate_2020"]]

#Change state codes to Names
unemployment_2020_df['State'] = unemployment_2020_df['State'].astype(str).apply(get_state_name)

#Group by State
unemployment_2020_df = unemployment_2020_df.groupby("State").mean()

unemployment_2020_df.head()
```

```python
#Find Columns for 2020 for Population
population_2020_df = combined_df.loc[:,["State","Area name","Population 2020"]]

#Change state codes to Names
population_2020_df['State'] = population_2020_df['State'].astype(str).apply(get_state_name)

#Group by State
population_2020_df = population_2020_df.groupby("State").mean()

population_2020_df.head()
```

```python
#Find Columns for 2020 for Poverty
poverty_2020_df = combined_df.loc[:,["State","Area name","Estimate of people of all ages in poverty 2020","Estimated percent of people of all ages in poverty 2020","Estimated percent of people age 0-17 in poverty 2020","Estimated percent of related children age 5-17 in families in poverty 2020","Estimate of median household income 2020","Estimated percent of children ages 0 to 4 in poverty 2020 (available for the U.S. and State total only)"]]

#Change state codes to Names
poverty_2020_df['State'] = poverty_2020_df['State'].astype(str).apply(get_state_name)

#Group by State
poverty_2020_df = poverty_2020_df.groupby("State").mean()

poverty_2020_df.head()
```

```python
#Create Dataframes for water quality

#Uranium
#import file
uranium_df = pd.read_csv("Resources\CDC\ANNUAL PERCENT OF CWS BY MEAN URANIUM CONCENTRAtion (MGL).csv")

#Rename Columns
uranium_df = uranium_df.rename(columns= {"Value":"Percent by Mean Uranium Concentration >30 mg/L"})
uranium_df.head()
```

```python
#Uranium
#import file
uranium_df = pd.read_csv("Resources\CDC\ANNUAL PERCENT OF CWS BY MEAN URANIUM CONCENTRAtion (MGL).csv")

#Rename Columns
uranium_df = uranium_df.rename(columns= {"Value":"Percent by Mean Uranium Concentration >30 mg/L"})
uranium_df.head()
```

```python
#arsenic
#import file
arsenic_df = pd.read_csv("Resources\CDC\ARSENIC IN COMMUNITY WATER SYSTEMS (CWS)  ANNUAL PERCENT OF CWS BY MEAN ARSENIC CONCENTation (MG per L).csv")

#Rename Columns
arsenic_df = arsenic_df.rename(columns= {"Value":"Percent by Mean Arsenic Concentration >10 mg/L"})
arsenic_df.head()
```

```python
#DEHP
#import file
dehp_df = pd.read_csv("Resources\CDC\DI(2-ETHYLHEXYL) PHTHALATE (DEHP).csv")

#Rename
dehp_df = dehp_df.rename(columns= {"Value":"Percent by Mean DI(2-ETHYLHEXYL) PHTHALATE (DEHP) Concentration > 6 mg/L"})
dehp_df.head()
```

```python
#Haloacetics Acids
#import file
hha_df = pd.read_csv("Resources\CDC\DISINFECTION BYPRODUCTS IN COMMUNITY WATER SYSTEMS (CWS)  ANNUAL PERCENT OF CWS BY MEAN Haloacetics Acids Concentration.csv")

#Rename Columns
hha_df = hha_df.rename(columns= {"Value":"Percent by Mean Haloacetics Acids Concentration >60 mg/L"})
hha_df.head()
```

```python
#Nitrate
#import file
nitrate_df = pd.read_csv(r"Resources\CDC\NITRATES IN COMMUNITY WATER SYSTEMS (CWS) ANNUAL PERCENT.csv")

#Rename Columns
nitrate_df = nitrate_df.rename(columns= {"Value":"Percent by Mean Nitrate Concentration >10 mg/L"})
nitrate_df.head()
```

```python
#Merge to create impurities_df
temp_df = pd.merge(uranium_df, arsenic_df, on="State", how="outer")
temp_df = pd.merge(temp_df, dehp_df, on="State", how="outer")
temp_df = pd.merge(temp_df, hha_df, on="State", how="outer")
temp_df = pd.merge(temp_df, nitrate_df, on="State", how="outer")
temp_df.head()
```

```python
#Create dataframe of impurities
impurities_df = temp_df.loc[:,["State","Year_x","Percent by Mean Uranium Concentration >30 mg/L","Percent by Mean Arsenic Concentration >10 mg/L","Percent by Mean DI(2-ETHYLHEXYL) PHTHALATE (DEHP) Concentration > 6 mg/L","Percent by Mean Nitrate Concentration >10 mg/L"]]

#Rename Columns
impurities_df = impurities_df.rename(columns={"Year_x":"Year"})

impurities_df.head()
```

```python
#Dictionary of States by Region
state_to_region = {
    'Arizona': 'Non-Coastal',
    'California': 'Pacific',
    'Colorado': 'Non-Coastal',
    'Connecticut': 'Atlantic',
    'Maine': 'Atlantic',
    'Maryland': 'Atlantic',
    'Massachusetts': 'Atlantic',
    "Michigan": "Non-Coastal",
    "Minnesota": "Non-Coastal",
    "Missouri": "Non-Coastal",
    'New Hampshire': 'Atlantic',
    'New Jersey': 'Atlantic',
    'New Mexico': 'Non-Coastal',
    'New York': 'Atlantic',
    'Oregon': 'Pacific',
    'Pennsylvania': 'Non-Coastal',
    'Rhode Island': 'Atlantic',
    'Utah': 'Non-Coastal',
    'Vermont': 'Non-Coastal',
    'Washington': 'Pacific',
    "Wisconsin": "Non-Coastal"
}

# Apply the mapping to the "State" column and create a new "Region" column
impurities_df['Region'] = impurities_df['State'].map(state_to_region)

#Group by Region
impurities_region_df = impurities_df.groupby("Region").mean()
impurities_region_df
```

```python
#Combine Census data together
#Dataframes to combine for the year 2020
census_2020_df = pd.merge(population_2020_df, unemployment_2020_df, on="State", how="outer")
census_2020_df = pd.merge(census_2020_df, education_2020_df, on="State", how="outer")
census_2020_df = pd.merge(census_2020_df, poverty_2020_df, on="State", how="outer")

#Reset index to create "State" column
census_2020_df.reset_index()

census_2020_df.info()
```

```python
#Merge impurities_df and Census 2020 dataframe
complete_2020_df =pd.merge(census_2020_df, impurities_df, on="State",how="outer")
complete_2020_df.head()
```

```python
#Write Complete dataframe to files

complete_2020_df.to_excel("complete_2020_df.xlsx")

complete_2020_df.to_csv("complete_2020_df.csv")
```
